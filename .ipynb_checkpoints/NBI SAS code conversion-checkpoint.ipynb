```python
# implemetn NOT part
# import some libs
import pandas as pd
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
import re
import string
import math
```

```python
# import main file (code lookup)
mainsas = pd.read_excel('7001479 Japan Local Protocol covariates code list formatted for SAS v1.0.xlsx', index_col=None, sheet_name='Covariates for PS-match')
```

```python
# data cleaning
mainsas['cov'] = mainsas['Covariate'].str.strip()
mainsas['cov'] = mainsas['cov'].str.lower()
mainsas['Code containing'] = mainsas['Code containing'].astype(str).str.replace('.', '')
```

```python
# import code requirement
goalsas1 = pd.read_excel('NBI covariates name .xlsx', index_col=None, sheet_name='Diabetes complications')
goalsas2 = pd.read_excel('NBI covariates name .xlsx', index_col=None, sheet_name='Sheet2')
goalsas3 = pd.read_excel('NBI covariates name .xlsx', index_col=None, sheet_name='Sheet1')
```

```python
# data cleaning
goalsas1['cov'] = goalsas1['Covariates in table shells '].str.strip()
goalsas1['cov'] = goalsas1['cov'].str.lower()
goalsas2['cov'] = goalsas2['Covariates in table shells '].str.strip()
goalsas2['cov'] = goalsas2['cov'].str.lower()
goalsas3['cov'] = goalsas3['Covariates in table shells '].str.strip()
goalsas3['cov'] = goalsas3['cov'].str.lower()
```

```python
# merge to get lookup info
mergesas1 = mainsas.merge(goalsas1, how='inner', left_on='cov', right_on='cov')
mergesas2 = mainsas.merge(goalsas2, how='inner', left_on='cov', right_on='cov')
mergesas3 = mainsas.merge(goalsas3, how='inner', left_on='cov', right_on='cov')
```

```python
# generate sas diseasecode
dc1 = mergesas1[mergesas1['Variable'].str.contains("Diseasecode")]
for i in dc1.itertuples():
    print('***', i[7],';')
    print('if', i[3], 'in', '(', i[4], ')', 'then', i[9]+'_dx', '= 1;')
    print('else', i[9]+'_dx', '= 0;')
    print()
print('dc1 end')

# generate sas disease code   
dc2 = mergesas2[mergesas2['Variable'].str.contains("Diseasecode")]
for i in dc2.itertuples():
    print('***', i[7],';')
    print('if', i[3], 'in', '(', i[4], ')', 'then', i[9]+'_dx', '= 1;')
    print('else', i[9]+'_dx', '= 0;')
    print()
print('dc2 end')

# generate sas disease code   
dc3 = mergesas3[mergesas3['Variable'].str.contains("Diseasecode")]
for i in dc3.itertuples():
    print('***', i[7],';')
    print('if', i[3], 'in', '(', i[4], ')', 'then', i[9]+'_dx', '= 1;')
    print('else', i[9]+'_dx', '= 0;')
    print()
print('dc3 end')
```

```python
# generate sas icd10 code for sheet 1
icd1 = mergesas1[mergesas1['Variable'].str.contains('ICD10code')]

def split(x, length):    
    splitted_string = x.split(',')
    result = [substr.strip().replace('.', '') for substr in splitted_string if len(substr.strip())==length]
    result = ', '.join(result)
    return result

for length in [5, 6, 7]:
    icd1['code_in%01d' % length] = icd1['Code containing'].apply(lambda x: split(x, length))
    icd1['code_ex%01d' % length] = icd1['Excluding'].apply(lambda x: split(x, length))
    
# generate sas icd10 code for sheet 2
icd2 = mergesas2[mergesas2['Variable'].str.contains('ICD10code')]

def split(x, length):    
    splitted_string = x.split(',')
    result = [substr.strip().replace('.', '') for substr in splitted_string if len(substr.strip())==length]
    result = ', '.join(result)
    return result

for length in [5, 6, 7]:
    icd2['code_in%01d' % length] = icd2['Code containing'].apply(lambda x: split(x, length))
    icd2['code_ex%01d' % length] = icd2['Excluding'].apply(lambda x: split(x, length))
    
def split_and_contain(x):    
    splitted_string = x.split(',')
    result = [substr.strip().replace('.', '') for substr in splitted_string if 'le' in substr.strip()]
    result = ', '.join(result)

    return result

icd2['test'] = icd2['Code containing'].apply(lambda x: split_and_contain(x))

# generate sas icd10 code for sheet 3
icd3 = mergesas3[mergesas3['Variable'].str.contains('ICD10code')]

def split(x, length):    
    splitted_string = x.split(',')
    result = [substr.strip().replace('.', '') for substr in splitted_string if len(substr.strip())==length]
    result = ', '.join(result)
    return result

for length in [5, 6, 7]:
    icd3['code_in%01d' % length] = icd3['Code containing'].apply(lambda x: split(x, length))
    icd3['code_ex%01d' % length] = icd3['Excluding'].apply(lambda x: split(x, length))
```
